In this code, we are going to try various combination of values for the hyperparameters of Logistic Regression algorithm for the given data "mobile_cleaned.csv", and the select the best model with highest accuracy.

The most importance hyperparameters of Logistic Regression are:
1.) Regularization strength - C
2.) Regularization type - penalty
3.) Solver
4.) Max Iterations - max-iter

C = Float value, default 1.0
We are going to consider values = 0.1, 1, 10

penalty = ‘l1’, ‘l2’, ‘elasticnet’, None

solver = {‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘sag’, ‘saga’}, default=’lbfgs’

max_iter = int, default 100
We are going to consider values = 100,200,300

```python
#importing the required libraries
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
#reading the data from the .csv file
df = pd.read_csv("mobile_cleaned.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#dividing the dataset into feautures and labels

df1 = df.values 

data = df1[:,0:39]
label = df1[:,39:]
```

```python
#splitting the data into training data and testing data
(tr_data,ts_data,tr_label,ts_label) = train_test_split(data,label,test_size = 0.25,random_state=42)
```

```python
C = [0.1,1,10]
penalty = ['l2', 'none']
solver = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']
max_iter = [100,200,300]
report = {}
```

```python
for i in C:
    for j in max_iter:
        for k in solver:
            for l in penalty:
                
                #as liblinear doesnt work when penalty is none
                if (k=="liblinear") &(l == 'none'):
                    continue
                    
                #similarly we could remove the required cases where the error occurs.As l1 doesnt support in many solvers, we didnt consider it
                model = LogisticRegression(C=i,penalty=l,solver=k,max_iter=j)
                model.fit(tr_data,tr_label)
                a=model.score(tr_data,tr_label)
                print(f"\nFor C = {i}\nmax_iter = {j}\nsolver = {k}\npenalty ={l}\n Tr data score = {a}\n Ts data score : ",end="")
                print(b:=model.score(ts_data,ts_label))
                dict1[(i,j)] = [a,b] 
        
                pred = model.predict(ts_data)
                rep = classification_report(ts_label,pred)
                print(rep)
                rep = classification_report(ts_label,pred,output_dict=True)['accuracy']
                report[(i,l,k,j)] = rep 
```

```python
report
```

```python
max_acc=max(report.values())
```

```python
def find_key_by_value(dictionary, value):
    temp = []
    for key, val in dictionary.items():
        if val == value:
            temp.append(key)
    return temp
key =  find_key_by_value(report,max_acc)
```
