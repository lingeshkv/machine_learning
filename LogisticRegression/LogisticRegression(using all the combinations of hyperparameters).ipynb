In this code, we are going to try various combination of values for the hyperparameters of Logistic Regression algorithm for the given data "mobile_cleaned.csv", and the select the best model with highest accuracy.

The most importance hyperparameters of Logistic Regression are:

1.) Regularization strength - C

2.) Regularization type - penalty

3.) Solver

4.) Max Iterations - max-iter

**C** = Float value, default 1.0. We are going to consider values = 0.01,0.1, 1, 10,100

**penalty** = ‘l1’, ‘l2’, ‘elasticnet’, None.  But for elasticnet, l1_ratio must be specified. To avoid the complexity, we avoid it.

**solver** = {‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘sag’, ‘saga’}, default=’lbfgs’

**max_iter** = int, default 100.  We are going to consider values = 100,200,300,500

```python
#importing the required libraries
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
#reading the data from the .csv file
df = pd.read_csv("/content/machine_learning/LogisticRegression/mobile_cleaned.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#dividing the dataset into feautures and labels

df1 = df.values 

data = df1[:,0:39]
label = df1[:,39:]
```

```python
#splitting the data into training data and testing data
(tr_data,ts_data,tr_label,ts_label) = train_test_split(data,label,test_size = 0.25,random_state=42)
```

```python
#defining lists with possible values of hyperparameters

C = [0.01,0.1,1,10,100]
penalty = ['l1','l2','none']
solver = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']
max_iter = [100,200,300,500]
report = {}
inv = []
```

```python
#looping through various combinations of values for hyperparameters

for i in C:
    for j in max_iter:
        for k in solver:
            for l in penalty:
                
                #some solver-penalty pairs are incompatible, so we pass those conditions
                if ((k in ['liblinear'] and l in ['none']) or (k in ['newton-cg', 'sag', 'lbfgs'] and l in ['l1'])):
                    inv.append([k,l])
                    continue
                
                #creating the model with particular hyperparameters
                model = LogisticRegression(C=i,penalty=l,solver=k,max_iter=j)
                
                #training the data using training dataset
                model.fit(tr_data,tr_label)
                
                #training and testing scores
                tr_score=model.score(tr_data,tr_label)
                ts_score=model.score(ts_data,ts_label)
                
                print(f"\nFor C = {i}\nmax_iter = {j}\nsolver = {k}\npenalty ={l}\n Tr data score = {tr_score}\n Ts data score = {ts_score}")
                
                #classification report
                pred = model.predict(ts_data)
                rep = classification_report(ts_label,pred)
                print(rep)
                
                #storing the hyperparameters-accuracy in a dictionary in terms of key-value respectively
                rep = classification_report(ts_label,pred,output_dict=True)['accuracy']
                report[(i,l,k,j)] = rep 
```

```python
report
```

```python
#searching for maximum accuracy in the report dictionary
max_acc=max(report.values())
```

```python
#finding the hyperparameters combination with maximum accuracy

def find_key_by_value(dictionary, value):
    temp = []
    for key, val in dictionary.items():
        if val == value:
            temp.append(key)
    return temp
key =  find_key_by_value(report,max_acc)
```

```python
print(f'The combinations of hyperparameters which gives the maximum accuracy are : {key}\n Accuracy = {max_acc}')
```

As we have considered the models **ONLY based on ACCURACY**, we could not guarantee that the model will be the best in all terms as there are other aspects of the model to consider(i.e., like precision, recall, f1-score)
