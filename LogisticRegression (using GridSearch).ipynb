```python
#importing the required libraries
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
#reading the data from the .csv file
df = pd.read_csv("/content/machine_learning/mobile_cleaned.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#dividing the dataset into feautures and labels

df1 = df.values 

data = df1[:,0:39]
label = df1[:,39:]
```

```python
#splitting the data into training data and testing data
(tr_data,ts_data,tr_label,ts_label) = train_test_split(data,label,test_size = 0.25,random_state=42)
```

```python
#defining lists with possible values of hyperparameters

param = {'C':[0.01,0.1,1,10,100],
'penalty' : ['l1','l2','none'],
'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
'max_iter' : [100,200,300,500]}
```

```python
# Create a logistic regression model
logreg = LogisticRegression()
```

```python
grid_search = GridSearchCV(logreg,param)
```

```python
grid_search.fit(tr_data,tr_label)
```

```python
print("Best parameters:", grid_search.best_params_)
```
