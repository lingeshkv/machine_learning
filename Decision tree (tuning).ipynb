In this code, we are going to try various combination of values for the hyperparameters of Decision Tree algorithm for the given data "pima-indians-diabetes.csv", and the select the best model with highest accuracy.In this code, we are 

The most important hyperparameters of Decision Tree algorithm are:
1.) criterion = {“gini”, “entropy”, “log_loss”}
2.) max_depth = int value. 
3.) max_features =int value

For finding the optimal values of the hyperparameters, we are going to compare the models in terms of its cross validation score. The one with higher cross validation score is considered more accurate and optimal.

```python
#importing required libraries

import numpy as np
import pandas as pd
import sklearn
from sklearn.metrics import classification_report as cr
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.model_selection import cross_val_score as cvs
```

```python
#reading the data from .csv file

df = pd.read_csv("pima-indians-diabetes.csv",names = ["f1","f2","f3","f4","f5","f6","f7","f8","label"])
```

```python
df.head()
```

```python
df.shape
```

```python
#splitting dataset into features and labels

df1 = df.values

data = df1[:,0:8]
labels = df1[:,8:]
```

```python
type(data)
```

```python
#train test split
(tr_data,test_data,tr_label,test_label) = train_test_split(df,labels,test_size = 0.25,random_state=42)
```

```python
#hyperparameter values

criterion = ['gini','entropy','log_loss']
max_depth = range(1,21)
max_features = range(1,data.shape[1]+1)
report = {}
```

```python
#cycling through various combinations of hyperparameters and creating the model.
for c in criterion:
    for md in max_depth:
        for mf in max_features:
            model = DTC(random_state = 42, criterion = c,max_depth = md, max_features = mf)
            
            #calculating the cross validation score, and storing it in a dictionary.
            cv_score = cvs(model,tr_data,tr_label,cv=5)
            report[(c,md,mf)] = np.mean(cv_score)
```

```python
print(report)
```

```python
max_acc = max(report.values())
```

```python
def find_key_by_value(dictionary, value):
    temp = []
    for key, val in dictionary.items():
        if val == value:
            temp.append(key)
    return temp
key =  find_key_by_value(report,max_acc)
```

```python
#printing all the combinations with max accuracy
print(key)
```
